# Time series Analysis

```python
# Example: Plotting time series for 'Fuel_and_light_data' in the 'Urban' sector
urban_Fuel_and_light_data = cpi[cpi['Sector'] == 'Urban']['Fuel and light']

plt.figure(figsize=(12, 6))
plt.plot(urban_Fuel_and_light_data, label='Fuel and light- Urban', color='blue')
plt.title('Fuel and light in Urban sector over time')
plt.xlabel('Month')
plt.ylabel('Fuel and light')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Example: Decomposing fuel and light in Urban
result = seasonal_decompose(cpi[cpi['Sector']=='Urban']['Fuel and light'], model='additive', period=12)
result.plot()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

# Assuming 'data' is your time series data
result = adfuller(cpi[cpi['Sector']=='Urban']['Fuel and light'])
# Extracting ADF statistic and p-value
adf_statistic = result[0]
p_value = result[1]

# Printing the results
print("ADF Statistic:", adf_statistic)
print("p-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("The time series is likely stationary.")
else:
    print("The time series is likely non-stationary.")
```

```python
# Assuming 'original_data' is your original time series data
# Specify the window size for the moving average
window_size = 12  # Choose an appropriate window size based on the frequency of your data (e.g., monthly, quarterly)

# Calculate the rolling mean (moving average)
rolling_mean = urban_Fuel_and_light_data .rolling(window=window_size).mean()

# Detrend the data by subtracting the rolling mean from the original data
detrended_data = urban_Fuel_and_light_data  - rolling_mean
```

```python
# Plot original data and detrended data
plt.figure(figsize=(12, 6))
plt.plot(urban_Fuel_and_light_data, label='Original Data', color='blue')
plt.plot(detrended_data, label='Detrended Data', color='red')
plt.title('Original Data vs. Detrended Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
